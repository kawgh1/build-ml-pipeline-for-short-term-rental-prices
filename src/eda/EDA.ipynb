```python
import wandb
import pandas as pd

# Load the dataset
# We use save_code=True so the notebook itself is uploaded and versioned in W&B.
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
local_path = wandb.use_artifact("sample.csv:latest").file()
df = pd.read_csv(local_path)
```

```python
# install setuptools and wheel in the current kernel
# !pip install ydata-profiling setuptools wheel
import ydata_profiling



# generate the profile report
profile = ydata_profiling.ProfileReport(df)
# profile.to_widgets() # Python 3.13 version bug
profile.to_file("profile_report.html")
```

```python
# Note: We are not imputing missing values here. 
# Missing value handling will be performed later in the inference pipeline 
# so that the system can manage them at prediction time.

# Drop outliers
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df = df[idx].copy()

# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```
